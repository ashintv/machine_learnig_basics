```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# load the digit dataset
digits = datasets.load_digits()
x = digits.data
y = digits.target

# printing the size of the dataset
print(f"Size of the dataset: {x.shape[0]} samples, {x.shape[1]} features")

# splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)

# initialise the SVM classifier
model = SVC(kernel='linear')

# train the model
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy is: {accuracy*100:.2f}%')

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion matrix:')
print(conf_matrix)

# Print the support vectors
print(f"Number of support vectors: {len(model.support_vectors_)}")
print("Support vectors:")
print(model.support_vectors_)
sp = model.support_vectors_

# Evaluation metrics
print('Accuracy score:', accuracy_score(y_test, y_pred))
p = precision_score(y_test, y_pred, average=None)
r = recall_score(y_test, y_pred, average=None)

for i in range(len(p)):
    print(f'class {i} precision={round(p[i]*100,2)}%   recall={round(r[i]*100,2)}%')

# Plot the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=digits.target_names)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
